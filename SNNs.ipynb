```python
# Load Pkgs
import pandas as pd
import networkx as nx

#
import warnings
```

```python
# und
g = nx.Graph()
```

```python
# Create or Add A Node to the Graph
g.add_node("Obama")
g.add_node("Trump")
```

```python
# Adding multiple nodes
g.add_nodes_from(["Hillary","Michelle"])
```

```python

g.nodes()
```

```python
g.add_edge("Obama","Michelle")
g.add_edge("Hillary","Clinton")
g.add_edge("Trump","Ivanka")
g.add_edge("Trump","Clinton")# Presidential Election
g.add_edge("Obama","Mitt")# Presidential Election
```

```python
### Adding Multiple relations
g.add_edges_from([("Hillary","Clinton"),("Obama","Trump"),("Obama","Clinton"),("Michelle","Ivanka"),("Michelle","Hillary")])
```

```python
g.nodes()
```

```python
# View Edges or Connections
g.edges()
```

```python
print("Number of nodes:", g.number_of_nodes())
print("Number of edges:", g.number_of_edges())
print("Is directed:", g.is_directed())
print("All nodes:", list(g.nodes))
print("All edges:", list(g.edges))
```

```python
# @title Visualization of Graph

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Method 1
# without labels by default
nx.draw(g)
```

```python
# Method 2
# with labels by setting it
nx.draw(g,with_labels=True)
```

```python
# Method 3
# Prints the names
nx.draw_networkx(g)
```

```python
### Using Plt
plt.figure(figsize=(10,10))
nx.draw_networkx(g)
plt.show()
```

```python
nx.spring_layout(g)
nx.draw_networkx(g)
```

```python
# @title Analysis

# How many people are connected to Obama
nx.degree(g,"Obama")
```

```python
nx.degree(g,"Ivanka")
```

```python
nx.degree(g,"Mitt")
```

```python
# What is the shortest connection between Obama and Bill Clinton
nx.shortest_path(g,"Obama","Clinton")
```

```python
# What is the shortest connection between Mitt and Trump
nx.shortest_path(g,"Mitt","Ivanka")
```

```python
# @title Degree Centrality

nx.degree_centrality(g)
```

```python
# @title Eigenvector Centrality
# degre=num of edge/num of no

nx.eigenvector_centrality(g)
```

```python
# @title Betweeness Centrality

nx.betweenness_centrality(g)
```

```python
# @title Predicting Associations
nx.clustering(g,"Obama")
```

```python
nx.average_clustering(g)
```

```python
# Reading our Edgelist
import pandas as pd
df = pd.read_csv("ga_edgelist.csv")
```

```python
df.head()
```

```python
greys = nx.from_pandas_edgelist(df,source="from",target="to")
greys
```

```python
print("Number of nodes:", greys.number_of_nodes())
print("Number of edges:", greys.number_of_edges())
print("Is directed:", greys.is_directed())
print("All nodes:", list(greys.nodes))
print("All edges:", list(greys.edges))
```

```python
# Check For All Nodes/People/Objects
greys.nodes
```

```python
# Check For All Edges /Connections
greys.edges
```

```python
# @title Visualization
nx.draw_networkx(greys)
```

```python
plt.figure(figsize=(10,10))
nx.draw_networkx(greys)
plt.show()
```

```python
# General
nx.degree(greys)
```

```python
# How many people are connected to Dr Grey
nx.degree(greys,"grey")
```

```python
# Neighbours
greys.neighbors("grey")
```

```python
# All Neighbours
nx.all_neighbors(greys,"grey")
```

```python
# @title Building A Subgroup

new_group = nx.bfs_tree(greys, "grey")
nx.draw_networkx(new_group)
```

```python
new_group.nodes
```
